## EDA Преобразование признаков. Нормализация. Стандартизация

````python
# создадим массив случайных чисел
np.random.seed(34)


#  Для нормализации, стандартизации используем следующую библиотеку
from sklearn import preprocessing

# Для графиков
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
matplotlib.style.use('ggplot')

# Сгенерируем датасет из случайных чисел 

df_random = pd.DataFrame({
    # Бета распределение, 5 – значение альфа, 1 – значение бета, 1000 – размер
    'beta': np.random.beta(5, 1, 1000) * 60,
    
    # Экспоненциальное распределение, 10 – "резкость" экспоненты, 1000 – размер
    'exponential': np.random.exponential(10, 1000),
    
    # Нормальное распределение, 10 – среднее значение р., 2 – стандартное отклонение, 1000 – количество сэмплов
    'normal_p': np.random.normal(10, 2, 1000),
    
    # Нормальное распределение, 10 – среднее значение р., 10 – стандартное отклонение, 1000 – количество сэмплов
    'normal_l': np.random.normal(10, 10, 1000)
})

# Копируем названия столбцов, которые теряются при использовании fit_transform()
col_names = list(df_random.columns)
df_random.head(3)
````

Сгенерированные распределения выбраны случайным образом, однако вы можете встретить их, например, в таких наборах данных:

* **Бета-распределение** моделирует вероятность. Например, коэффициент конверсии клиентов, купивших что-то на сайте.
* **Экспоненциальное распределение**, предсказывающее периоды времени между событиями. Например, время ожидания автобуса.
* **Нормальное распределение**, например распределение роста и веса человека.

````python
# зададим параметры холста, название и визуализируем кривые распределения:
fig, (ax1) = plt.subplots(ncols=1, figsize=(10, 8))
ax1.set_title('Исходные распределения')

# kdeplot() (KDE – оценка плотности ядра) – специальный метод для графиков распределений
sns.kdeplot(df_random['beta'], ax=ax1, label ='beta')
sns.kdeplot(df_random['exponential'], ax=ax1, label ='exponential')
sns.kdeplot(df_random['normal_p'], ax=ax1, label ='normal_p')
sns.kdeplot(df_random['normal_l'], ax=ax1, label ='normal_l')
plt.legend()
````

````python
df_random.describe()
````

### MinMaxScaler

````python
# Инициализируем нормализатор
mm_scaler = preprocessing.MinMaxScaler()

# Кодируем исходный датасет
df_mm = mm_scaler.fit_transform(df_random)

# Преобразуем промежуточный датасет в полноценный датафрейм для визуализации

df_mm = pd.DataFrame(df_mm, columns=col_names)

fig, (ax1) = plt.subplots(ncols=1, figsize= (10, 8))
ax1.set_title('После нормализации MinMaxScaler')

sns.kdeplot(df_mm['beta'], ax=ax1)
sns.kdeplot(df_mm['exponential'], ax=ax1)
sns.kdeplot(df_mm['normal_p'], ax=ax1)
sns.kdeplot(df_mm['normal_l'], ax=ax1)
````

### EDA Проверка знаний

````python
df_heart = pd.read_csv('data/heart.zip')
df_heart.head()
````

````python
# Создадим признак old, где 1 - пациент больше 60
df_heart['old'] = df_heart['age'].apply(lambda x: 1 if x > 60 else 0)
df_heart['old'].sum()
````

````python
# Создадим новый признак с нормальным давлением по полу и возрасту trestbps_mean
def norm(col):
    if col['sex'] == 1 and col['age']<=20:
        return int(123)
    if col['sex'] == 0 and col['age']<=20:
        return int(116)
    if col['sex'] == 1 and 21 <= col['age'] <= 30:
        return int(126)
    if col['sex'] == 0 and 21 <= col['age'] <= 30:
        return int(120)
    if col['sex'] == 1 and 31 <= col['age'] <= 40:
        return int(129)
    if col['sex'] == 0 and 31 <= col['age'] <= 40:
        return int(127)
    if col['sex'] == 1 and 41 <= col['age'] <= 50:
        return int(135)
    if col['sex'] == 0 and 41 <= col['age'] <= 50:
        return int(137)
    if col['sex'] == 1 and 51 <= col['age'] <= 60:
        return int(142)
    if col['sex'] == 0 and 51 <= col['age'] <= 60:
        return int(144)
    if col['sex'] == 1 and col['age'] >= 61:
        return int(142)
    if col['sex'] == 0 and col['age'] >= 61:
        return int(159)
    
df_heart['trestbps_mean'] = df_heart.apply(norm, axis=1)
df_heart['trestbps_mean'][300]
````

````python
cp_encoder = ce.OneHotEncoder(cols=['cp']) # указываем столбец для кодирования
cp_bin = cp_encoder.fit_transform(df_heart['cp'])
df_heart = pd.concat([df_heart, cp_bin], axis=1)

restecg_encoder = ce.OneHotEncoder(cols=['restecg']) # указываем столбец для кодирования
restecg_bin = restecg_encoder.fit_transform(df_heart['restecg'])
df_heart = pd.concat([df_heart, restecg_bin], axis=1)

slope_encoder = ce.OneHotEncoder(cols=['slope']) # указываем столбец для кодирования
slope_bin = slope_encoder.fit_transform(df_heart['slope'])
df_heart = pd.concat([df_heart, slope_bin], axis=1)

ca_encoder = ce.OneHotEncoder(cols=['ca']) # указываем столбец для кодирования
ca_bin = ca_encoder.fit_transform(df_heart['ca'])
df_heart = pd.concat([df_heart, ca_bin], axis=1)

thal_encoder = ce.OneHotEncoder(cols=['thal']) # указываем столбец для кодирования
thal_bin = thal_encoder.fit_transform(df_heart['thal'])
df_heart = pd.concat([df_heart, thal_bin], axis=1)

df_heart.head()
````
