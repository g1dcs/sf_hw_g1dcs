# EDA Проектирование признаков

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/wine_cleared-1.csv')
df.head(5)
```

```python
# Преобразуем признак цены округлив его до целого и приведем к числовому типу
df['price_round'] = df['price'].round().astype(int)
df.head(5)
```

```python
reg = '\d{4}' # Регулятор для поиска цифр из 4 символов в строке
df['year'] = df['title'].str.findall(reg).str.get(0)
df.head(5)
```

```python
df['is_usa'] = df['country'].apply(lambda x: 1 if x == 'US' else 0)
df.head()
```

```python
df['country'].value_counts()
```

```python
df['is_france'] = df['country'].apply(lambda x: 1 if x == 'France' else 0)
df['is_italy'] = df['country'].apply(lambda x: 1 if x == 'Italy' else 0)
```

```python
print(df['is_france'].sum())
print(df['is_italy'].sum())
```

```python
# преобразуем признак year в объект datetime для удобного сравнения дат
df['year'] = pd.to_datetime(df['year'], errors='coerce')

# для сравнения используем год, заполняем значения признака old_wine, где год вина меньше 2010
df['old_wine'] = df['year'].apply(lambda x: 1 if x.year < 2010 else 0) 
df['old_wine'].sum()
```

```python
regex = '\((.*?)\)'
df['locality'] = df['title'].str.findall(regex).str.get(0)
df.info()

```

## EDA Работа с форматом "Дата-Время"

```python
# Создадим DF напримере кол центра, чтобы на практике попробовать определить стоит ли звонить определенному клиенту сейчас
calls_list = [
    [460, '2013-12-17 04:55:39', '2013-12-17 04:55:44', '2013-12-17 04:55:45'],
    [12, '2013-12-16 20:03:20', '2013-12-16 20:03:22', '2013-12-16 20:07:13'],
    [56, '2013-12-16 20:03:20', '2013-12-16 20:03:20', '2013-12-16 20:05:04'],
    [980, '2013-12-16 20:03:20','2013-12-16 20:03:27', '2013-12-16 20:03:29'],
    [396, '2013-12-16 20:08:27', '2013-12-16 20:08:28','2013-12-16 20:12:03'],
    [449, '2013-12-16 20:03:20', '2013-12-16 20:03:25','2013-12-16 20:05:00'],
    [397, '2013-12-16 20:08:25', '2013-12-16 20:08:27', '2013-12-16 20:09:59'],
    [398, '2013-12-16 20:01:23', '2013-12-16 20:01:23', '2013-12-16 20:04:58'],
    [452, '2013-12-16 20:03:20', '2013-12-16 20:03:21','2013-12-16 20:04:55'],
    [440, '2013-12-16 20:03:20', '2013-12-16 20:04:26', '2013-12-16 20:04:32']
]

calls = pd.DataFrame(calls_list, columns = ['client_id',  'agent_date', 'created_at' ,'end_date'])

# Преобразуем признаки в формат "DateTime"

calls['agent_date'] = pd.to_datetime(calls['agent_date'])
calls['created_at'] = pd.to_datetime(calls['created_at'])
calls['end_date'] = pd.to_datetime(calls['end_date'])

calls
```

```python
# Создадим признак длительности разговора из разницы конца разговора и начала

calls['duration'] = (calls['end_date'] - calls['created_at']).dt.seconds

calls
```

```python
# Создадим признак время соеденения

calls['time_connection'] = (calls['created_at'] - calls['agent_date']).dt.seconds

calls['time_connection'].sum()
```

```python
calls
```

```python
# Создадим признак состоялся ли звонок, при условии если разговор был более 10 секунд
calls['is_connection'] = calls['duration'].apply(lambda x: 1 if x > 10 else 0)
calls['is_connection'].sum()
```

```python
calls['time_diff'] = (calls['end_date'] - calls['agent_date']).dt.seconds
calls['time_diff'].sum()
```

```python
# Удалим признаки дат, так как модели работают только с числовыми значениями
calls = calls.drop(columns=['agent_date', 'created_at' ,'end_date'], axis=1)
```
