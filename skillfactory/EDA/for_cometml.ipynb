## Кодировка категориальных признаков

```python
categorical_columns_names = ['Gender', 'Neighbourhood'] # выбираем те колонки которые нужно перекодировать из объекта в числовые значения
```

```python
encoded_columns = pd.get_dummies(df, columns = categorical_columns_names) # get_dummies это альтернатива OneHotEncoding
```

```python
encoded_columns.drop(['AppointmentID', 'PatientId', 'ScheduledDay', 'AppointmentDay', 'No-show'], axis=1, inplace = True) # удаляем лишние столбцы
```

```python
encoded_columns.head()
```

```python
y = df['No-show'] # проверим столбец не пришедших клиентов
```

```python
y
```

```python
y = y.replace({'No': 0, 'Yes': 1}) # заменим слова yes/no на числовые значения для модели
```

```python
y.value_counts(True)
```

## Шкалирование данных

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(encoded_columns)
```

```python
print(X.shape)
```

# Модель (этот раздел будет подробно изучен далее в курсе)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### Инициализация модели

```python
logreg = LogisticRegression()
```

```python
random_state = 42 # для воспроизводимости результатов, как с randit 
```

### Обучение модели

```python
logreg.fit(X, y)
```

#### Логирование параметров обучения

```python
params={"random_state":random_state,
        "model_type":"logreg"
}
```

```python
experiment.log_parameters(params)
```

### Предсказания модели

```python
y_pred = logreg.predict(X)  # создаем предсказания при помоши метода pandas predict
```

### Анализ модели

```python
accuracy = accuracy_score(y, y_pred) # сравниваем реальные параметры с предсказанными

print("Доля правильных ответов: {:6.3f}".format(accuracy))
```

# Логирование метрик

```python
metrics = {"accuracy":accuracy}
```

```python
experiment.log_metrics(metrics)
```
