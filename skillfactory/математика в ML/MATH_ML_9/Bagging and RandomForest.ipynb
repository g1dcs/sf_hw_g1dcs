# Задания к модулю 2


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import plotly.express as px
import numpy as np
from sklearn import tree
from sklearn import model_selection
from sklearn import metrics
from sklearn import linear_model 
from sklearn import ensemble
```

```python
df = pd.read_csv('Data/wineQualityReds.csv', sep=',')
df
```

```python

df['target']= np.where(df.quality >=6,1,0)
Y= df['target']
X = df.drop(['target', 'quality'], axis= 1)
X_train, X_test, y_train, y_test = model_selection.train_test_split(
    X, Y, test_size=0.3, random_state=42)
print(X.shape, Y.shape)
```

```python
lr = linear_model.LogisticRegression()
lr.fit(X_train, y_train)

# Делаем предсказания для тестовой и тренировочной выборки
y_train_pred_lr = lr.predict(X_train)
y_test_pred_lr = lr.predict(X_test)

# Проверим метрики 
print(f'Значение f1 на тестовой выборке lr: {metrics.f1_score(y_test, y_test_pred_lr):.3f}')
print(f'Значение f1 на тренеровочной выборке lr: {metrics.f1_score(y_train, y_train_pred_lr):.3f}')

dt = tree.DecisionTreeClassifier(
    max_depth=10,
    random_state=42
)
dt.fit(X_train, y_train)

# Делаем предсказания для тестовой и тренировочной выборки
y_train_pred_dt = dt.predict(X_train)
y_test_pred_dt = dt.predict(X_test)

# Проверим метрики 
print(f'Значение f1 на тестовой выборке dt : {metrics.f1_score(y_test, y_test_pred_dt):.3f}')
print(f'Значение f1 на тренеровочной выборке dt: {metrics.f1_score(y_train, y_train_pred_dt):.3f}')
```

```python
#Создаем объект класса случайный лес
bc = ensemble.BaggingClassifier(
    estimator=dt,
    n_estimators=1500,
    random_state=42
)
#Обучаем модель
bc.fit(X_train, y_train)
#Выводим значения метрики 
y_train_pred = bc.predict(X_train)
print('Train: {:.3f}'.format(metrics.f1_score(y_train, y_train_pred)))
y_test_pred = bc.predict(X_test)
print('Test: {:.3f}'.format(metrics.f1_score(y_test, y_test_pred)))
```

# Задания к модулю RandomForest


```python
data = pd.read_csv('data/boston.csv', decimal=",")
data.head()
```

```python
y = data['target']
X = data.drop(['target'], axis= 1)
X_train, X_test, y_train, y_test = model_selection.train_test_split(
    X, y, test_size=0.3, random_state=13)

print(X.shape, y.shape)
```

```python
y_train.mean()
```

```python
lr = linear_model.LinearRegression()
lr.fit(X_train, y_train)

y_train_pred_lr = lr.predict(X_train)
print('Train MAE: {:.2f}'.format(metrics.mean_absolute_error(y_train, y_train_pred_lr)))
y_test_pred_lr = lr.predict(X_test)
print('Test MAE: {:.2f}'.format(metrics.mean_absolute_error(y_test, y_test_pred_lr)))
```

```python
rf = ensemble.RandomForestRegressor(random_state=13)
rf.fit(X_train, y_train)

y_test_pred_rf = rf.predict(X_test)
print('Test MAE: {:.2f}'.format(metrics.mean_absolute_error(y_test, y_test_pred_rf)))
```

```python
rf = ensemble.RandomForestRegressor(
    n_estimators=3,
    random_state=13
)
rf.fit(X_train, y_train)

y_test_pred_rf = rf.predict(X_test)
print('Test MAE: {:.2f}'.format(metrics.mean_absolute_error(y_test, y_test_pred_rf)))


rf = ensemble.RandomForestRegressor(
    n_estimators=10,
    random_state=13
)
rf.fit(X_train, y_train)

y_test_pred_rf = rf.predict(X_test)
print('Test MAE: {:.2f}'.format(metrics.mean_absolute_error(y_test, y_test_pred_rf)))


rf = ensemble.RandomForestRegressor(
    n_estimators=100,
    random_state=13
)
rf.fit(X_train, y_train)

y_test_pred_rf = rf.predict(X_test)
print('Test MAE: {:.2f}'.format(metrics.mean_absolute_error(y_test, y_test_pred_rf)))

rf = ensemble.RandomForestRegressor(
    n_estimators=500,
    random_state=13
)
rf.fit(X_train, y_train)

y_test_pred_rf = rf.predict(X_test)
print('Test MAE: {:.2f}'.format(metrics.mean_absolute_error(y_test, y_test_pred_rf)))
```
