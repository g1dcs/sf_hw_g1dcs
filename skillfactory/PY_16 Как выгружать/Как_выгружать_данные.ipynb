### Решаем проблему с кодировкой исходных данных (когда не правильно читаются символы)

```python
# Н.р. считаем фаил без решения проблемы
data=pd.read_csv('data/ErrorEnCoding.csv', header=None, encoding='latin-1')
display(data)
```

```python
# Для коректного чтения нужно определить кодировку
from chardet.universaldetector import UniversalDetector

detector = UniversalDetector()
with open('data/ErrorEnCoding.csv', 'rb') as fh:
    for line in fh:
        detector.feed(line)
        if detector.done:
            break
detector.close()
# с увереностью в 80% тип кодировки koi8-r
```

```python
# Считаем файл указав кодировку
data = pd.read_csv('data/ErrorEnCoding.csv', encoding='koi8-r', header=None)
data
```

### Немного об Excel

```python
# Для чтени фаило excel исп функция pandas read_excel()
data = pd.read_excel('data/grades.xlsx')
display(data)
```

```python
data = pd.read_excel('https://github.com/asaydn/test/raw/master/january.xlsx')
display(data)
```

```python
# Чтобы считать информацию со 2 листа нужно использовать параметр sheet_name=
data = pd.read_excel('data/grades.xlsx', sheet_name='ML')
display(data)
```

```python
# Для сохранения DF в excel используется метод to_excel()
data.to_excel('data/grades_new.xlsx')

# В этом случае будет создан один лист с именем по умолчанию "Sheet1". 
# Также мы сохраним и индекс — в данных будет находиться лишний столбец. 
# Чтобы создать лист с определённым именем (например, Example) и не сохранять индекс,
# в метод  to_excel() необходимо передать параметры sheet_name='Example' 

data.to_excel('data/grades_new.xlsx', index=False, sheet_name='Example')
```

```python
df_ratings = pd.read_excel('data/ratings+movies.xlsx', sheet_name='ratings')
display(df_ratings)
```

```python
df_movie = pd.read_excel('data/ratings+movies.xlsx', sheet_name='movies')
display(df_movie)
```

```python
merged = df_movie.merge(
    df_ratings,
    on='movieId',
    how='right'
)
display(merged)
```

```python
merged.to_excel('data/ratings+movie_new.xlsx', index=False, sheet_name='Example')
```
