### 16.6 JSON. Работа в Pandas, из JSON в PANDAS

```python
import json
from pprint import pprint
import pandas as pd


with open('recipes.json') as f:
    recipes = json.load(f)
df = pd.DataFrame(recipes)
display(df.head())
```

```python
# Или можно сразу читать через возможности Pandas
df_1 = pd.read_json('recipes.json')
display(df_1.head())
```

```python
# Создадим множество из игридиентов
with open('recipes.json') as f:
    recipes = json.load(f)
all_ingredients = set()
for recipe in recipes:
    for ingredient in recipe['ingredients']:
        all_ingredients.add(ingredient)
display(len(all_ingredients))

df_recept = pd.DataFrame(recipes)

```

Теперь определим функцию contains(), с помощью которой мы будем проверять наличие конкретного ингредиента ingredient_name в рецепте текущего блюда, который представлен списком ingredient_list (значение в ячейке столбца ingredients текущего рецепта).

```python
def contains(ingredient_list):
    if ingredient_name in ingredient_list:
        return 1
    else:
        return 0
```

```python
for ingredient_name in all_ingredients:
    df_recept[ingredient_name] = df_recept['ingredients'].apply(contains)
df_recept['ingredients'] = df_recept['ingredients'].apply(len)
display(df_recept.head())
```

```python
df_recept.to_csv('recipes.csv', index=False)
```

### 16.7 Обратная выгрузка из Pandas в JSON

```python
import pandas as pd
df = pd.read_csv('recipes.csv')
ids = list(df['id'].unique())
ingredients = list(df.columns)[3:]

def make_list(row):
    ingredients_list = []
    for ingredient in ingredients:
        if row[ingredient].item() == 1:
            ingredients_list.append(ingredient)
    return ingredients_list

new_recipes = [] # Создаём пустой список для хранения итоговой структуры
for current_id in ids: # Организуем цикл с параметром current_id
    cuisine = df[df['id'] == current_id]['cuisine'].iloc[0] # Получаем значение соответствующей кухни, применив фильтр по текущему значению параметра цикла к DataFrame;
    current_ingredients = make_list(df[df['id'] == current_id]) # Получаем перечень ингредиентов, входящих в состав текущего блюда
    current_recipe = {'cuisine': cuisine, 'id': int(current_id), 'ingredients': current_ingredients} # Создаём текущий словарь
    new_recipes.append(current_recipe) # Добавляем созданный словарь к списку
```

```python
pprint(new_recipes)
```

```python
new_recipes = json.dumps(new_recipes) # Функция dumps() модуля json сериализирует объект Python в строку формата JSON. 
with open("data/new_recipes.json", "w") as write_file: # Откроем файл new_recipes.json для записи
    write_file.write(new_recipes) # Записываем содержимое подготовленные данные в файл
```

## 16.9 XML контент

```python
# для загрузки используем встроеннную библиотеку
import xml.etree.ElementTree as ET
tree = ET.parse('data/menu.xml')
```

```python
# Запишем в переменную root корневой узел дерева tree и 
# посмотрим, как выглядит содержимое переменной root, для чего выполним код
root = tree.getroot()
display(root)
# Проверим тип переменной
display(type(root))
```

```python
# Для того чтобы посмотреть список потомков корневого узла, выполним следующий код
display(list(root))
```

```python
# Что бы посмотреть потомков dish выведем следующий код
display(list(root[1]))
```

```python
# У узлов могут быть атрибуты н.р. имя, чтобы их увидеть воспользуемся следующим кодом
display(root[0].attrib)
```

```python
# Для обращения к узловому показателю необходимо использовать атрибут text
display(root[0][0])
display(root[0][0].text)
```

```python
# Для прочтения конкретного узла нужно использовать параметр teg
display(root.tag)
display(root[0][2].tag)
```

```python
# Для преобразования XML создадим цикл
for dish in root:
    for param in dish:
        print(dish.attrib['name'], param.tag, param.text)
    print()
```
